## Data Wrangling

```python
filename = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

data = pd.read_csv(filename, names = headers)
data.head()
```

### Missing Values

```python
# Replace "?" to NaN,
data.replace("?", np.nan, inplace = True)
data.head(5)
```

```python
missing_data = data.isnull()
missing_data.head(5)
```

```python
for col in missing_data.columns.values.tolist():
    print(col)
    print(missing_data[col].value_counts())
    print("")    
```

### Deal with missing data,

- Drop data,

 1. drop the whole row

 2. drop the whole column
- Replace data,

 1. replace it by mean

 2. replace it by frequency

 3. replace it based on other functions

Whole columns should be dropped only if most entries in the column are empty. In our dataset, none of the columns are empty enough to drop entirely:

Replace by mean:

"normalized-losses": 41 missing data, replace them with mean
"stroke": 4 missing data, replace them with mean
"bore": 4 missing data, replace them with mean
"horsepower": 2 missing data, replace them with mean
"peak-rpm": 2 missing data, replace them with mean
Replace by frequency:

"num-of-doors": 2 missing data, replace them with "four".
Reason: 84% sedans is four doors. Since four doors is most frequent, it is most likely to occur
Drop the whole row:

"price": 4 missing data, simply delete the whole row
Reason: price is what we want to predict. Any data entry without price data cannot be used for prediction; therefore any row now without price data is not useful to us

```python
avg_norm_loss = data["normalized-losses"].astype("float").mean(axis=0)
print("The average of normalized-losses:", avg_norm_loss)
```

```python
data["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore = data["bore"].astype("float").mean(axis=0)
print("The average of Bore Values:",avg_bore)
```

```python
data["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = data["stroke"].astype("float").mean(axis=0)
print("The average of Stroke Values:",avg_bore)
```

```python
data["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_horsepower = data['horsepower'].astype("float").mean(axis=0)
print("The average horsepower:", avg_horsepower)
```

```python
data['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm= data['peak-rpm'].astype('float').mean(axis=0)
print("The average peak rpm:", avg_peakrpm)
```

```python
data['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# replace the missing values in 'num-of-doors' by the most frequent, 
data['num-of-doors'].value_counts()
```

```python
data['num-of-doors'].value_counts().idxmax()
```

```python
data['num-of-doors'].replace(np.nan, "four", inplace=True)
```

```python
# drop whole row with NaN in "price" column,
data.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
data.reset_index(drop=True, inplace=True)
```

##### Looking at the datset, some columns are not of the correct data type. Numerical variables should have type 'float' or 'int', and variables with strings such as categories should have type 'object'. For example, 'bore' and 'stroke' variables are numerical values that describe the engines, so we should expect them to be of the type 'float' or 'int'; however, they are shown as type 'object'. We have to convert data types into a proper format for each column using the "astype()" method.

```python
#Converting the datatypes,
data[["bore", "stroke"]] = data[["bore", "stroke"]].astype("float")
data[["normalized-losses"]] = data[["normalized-losses"]].astype("int")
data[["price"]] = data[["price"]].astype("float")
data[["peak-rpm"]] = data[["peak-rpm"]].astype("float")
```

### Data Standardization
Data is usually collected from different firms with different formats. (Data Standardization is also a term for a particular type of data normalization, where we subtract the mean and divide by the standard deviation)

```python
data.head()
```

```python
# 1 Litre per 100kms = 235/ mpg (US)
data["city-L/100km"] = 235 / data["city-mpg"]
data.head()
```

```python
# 1 Litre per 100kms = 235/ mpg (US)
data["highway-mpg"] = 235/data["highway-mpg"]
data.head()
```

### Data Normalization:

- Why normalization?

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, or scaling variable so the variable values range from 0 to 1.

```python
# Therefore, replace (original value) by (original value)/(maximum value)
data["length"] = data["length"]/data["length"].max()
data["width"] = data["width"]/data["width"].max()
data['height'] = data['height']/data['height'].max() 
data[["length","width","height"]].head()
```

We can see that, we've normalized "length", "width" and "height" in the range of 0 to 1.

### Binning:

- Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.

```python
data["horsepower"] = data["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(data["horsepower"])

plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("HORSEPOWER BINS")
```

```python
bins = np.linspace(min(data["horsepower"]), max(data["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
#We apply the function "cut" the determine what each value of "df['horsepower']" belongs to.
data['horsepower-binned'] = pd.cut(data['horsepower'], bins, labels= group_names ,include_lowest = True)
data[['horsepower','horsepower-binned']].head(20)
```

```python
data["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, data["horsepower-binned"].value_counts())

# labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

Now looking at the dataset we see that teh intervals have narrowed from 57 to 3 categories!("Low","Medium" and "High")

```python
# BIN VISUALIZATION,

# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(data["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

#### Dummy variable:
- a numerical variable used to label categories called 'dummies' because the numbers themselves don't have inherent meaning.

```python
data.columns
```

```python
dummy_variable_1 = pd.get_dummies(data["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

Now in the column "fuel-type", 
- value 0 = "gas" 
- value 1 = "diesel"  

```python
# merging 'dummy_variable_1' column into our original dataset 'data'.
df = pd.concat([data, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "data"
df.drop("fuel-type", axis = 1, inplace=True)
```
